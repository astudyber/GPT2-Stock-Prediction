# 4. 指定的数据格式处理

```python
'''
[
	{
		'day1':'1',
		'day2':'1 2',
		'day10':'1 2 … 10',
		...
		'day1close1':'2',
		'day1close5':'2 3 4 5 6',
	},
    {
    ...
    },
    ...
]
'''
# 最终的数据
all_data = []




# 遍历每个 csv-str 文件
for i,v in enumerate(data_list_str):
    stock_name = file_names[i]

    # 倒序！！！！！！！
    v.reverse()
    data_list[i] = data_list[i].iloc[::-1]

    print(stock_name+' 处理中...')
    # 生成每个字典
    for j in range(len(v)-14):
        one_data = {}
        for k in range(10):  # 用十天数据
            key_name = 'day'+str(k+1)
            one_data[key_name] = stock_name +' '+ v[j]
            for item in range(k):
                one_data[key_name] = one_data[key_name] +' '+ v[j+item+1]
            for l in range(5):  # 向后预测五天的收盘价
                key_name2 = 'day'+str(k+1) + 'close'+str(l+1)
                one_data[key_name2] = stock_name +' '+ str(data_list[i].iloc[j+k+1]['close'])
                for item2 in range(l):
                    one_data[key_name2] = one_data[key_name2] +' '+ str(data_list[i].iloc[j+k+2+item2]['close'])
                # print(one_data[key_name2])
        
        all_data.append(one_data)

    #     break
    # break

    # 将数据保存到本地json文件中
    with open('dataset/' + str(i) +'.json', 'w') as json_file:
        json.dump(all_data, json_file)
    all_data = []

# print(all_data)
```
